```julia
using PyPlot, LinearAlgebra
```

# 単回帰

```julia
x=[-1 2 3]
y=[1 4 2]
```

```julia
b,a = inv([length(x) sum(x); sum(x) sum(x.*x)])*[sum(y); sum(x.*y)]
```

```julia
xx = range(-3, 3, length=10)
yy = a*xx + range(b, b, length=10)
plot(x, y, "o", label="data")
plot(xx, yy, "o", label="linear")
```

# 重回帰

```julia
x=[1 1 2; 1 -1 1; 1 3 0; 1 -2 -2]
y=[4; 2; 1; -1]
```

```julia
w = inv(x'*x)*x'*y
```

```julia
xx1 = range(-3, 3, length=10)
xx2 = range(-3, 3, length=10)
yy = range(w[1], w[1], length=10) + w[2]*xx1 + w[3]*xx2
```

```julia
plot(x[:,2], x[:,3], y, "o", label="data")
```

```julia
scatter(x[:,2], x[:,3], y)
```

```julia
surf(x[:,2], x[:,3], y)
surf(x[:,2], x[:,3], yy)
```
